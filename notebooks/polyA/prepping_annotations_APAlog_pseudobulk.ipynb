```python
import scanpy as sc
```

```python
adata_miR = sc.read('/home/ssobti/projects/mir_tud/101423_filtered_data/miR.integrated.qc_norm_scaled.dimred.clustered.gex.guide_assigned.h5ad')
adata_tRF = sc.read('/home/ssobti/projects/mir_tud/101423_filtered_data/tRF.integrated.qc_norm_scaled.dimred.clustered.gex.guide_assigned.h5ad')
```

```python
adata_miR_meta = adata_miR.obs[['barcode', 'sample']].copy()
adata_tRF_meta = adata_tRF.obs[['barcode', 'sample']].copy()
adata_miR_meta['guide'] = ['TuD_NC' if x == 'NA' else x for x in adata_miR.obs['miR.family']]
adata_tRF_meta['guide'] = ['TuD_NC' if x.startswith('TuD_NC') else x for x in adata_tRF.obs['guide']]
```

```python
adata_miR_meta.to_csv('/home/aiden/APA_trf/data/miR_annotations.csv', index=False)
adata_tRF_meta.to_csv('/home/aiden/APA_trf/data/tRF_annotations.csv', index=False)
```
