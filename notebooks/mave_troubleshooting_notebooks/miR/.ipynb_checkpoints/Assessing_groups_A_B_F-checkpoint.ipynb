```r
library(ggplot2)
library(ggpubr)
```

```r
file_list = list.files(path = '~/projects/mir_tud/output_data/mave_output/mave_trblsht/', pattern = '*matrix.tsv.gz' )
```

```r
file_list = file_list[-17]
```

```r
file_list
```

```r
length(file_list)
```

```r
file_paths = paste0('~/projects/mir_tud/output_data/mave_output/mave_trblsht/', file_list)
```

```r
latent_space_mtxs = lapply(X = file_paths, FUN = read.csv, sep = '\t')
```

```r
meta_weight_pattern_names = c('default', 'high*_reconstruction', 'high*_reconstruction_low*_triplet', 'high_KLD', 
                              'high_KLD_low_triplet', 'high*_triplet', 'high_triplet', 'low_triplet', 'low*_triplet', 
                              'low_KLD_high_triplet', 'low_KLD', 'all_even_high*_triplet', 'all_even', 'all_even_low*_triplet',
                              'low*_reconstruction_high*triplet', 'low*_reconstruction') 
```

```r
length(meta_weight_pattern_names)
```

```r
plotting_function = function(data, nm){ggplot(data, aes(LV.mu.2, LV.mu.3)) + geom_point() + coord_cartesian(xlim = c(-4,4), ylim = c(-4,4)) + ggtitle(nm)}
```

```r
plots = mapply(FUN = plotting_function, latent_space_mtxs, meta_weight_pattern_names, SIMPLIFY = FALSE)
```

```r
names(plots) = meta_weight_pattern_names
```

```r
plots = plots[c('default', 'low_KLD', 'high_KLD', 'low_triplet', 'high_triplet', 
                'low_KLD_high_triplet', 'high_KLD_low_triplet', 'low*_reconstruction', 'high*_reconstruction', 
               'low*_reconstruction_high*triplet', 'high*_reconstruction_low*_triplet', 'low*_triplet', 'high*_triplet',
                'all_even', 'all_even_low*_triplet', 'all_even_high*_triplet')]
```

```r
plotted_plots = plots[c('default', 'low_KLD', 'high_KLD', 'low*_triplet', 'high*_triplet',
                'low_KLD_high_triplet', 'high_KLD_low_triplet', 'low*_reconstruction', 'high*_reconstruction', 
               'low*_reconstruction_high*triplet', 'high*_reconstruction_low*_triplet',
                'all_even', 'all_even_low*_triplet', 'all_even_high*_triplet')]
```

```r
arranged_plots = ggarrange(plotlist = plotted_plots, legend = 'none', nrow = 3, ncol = 5, common.legend = TRUE)
```

```r
options(repr.plot.width=16, repr.plot.height=10)
arranged_plots

## remove KLD loss to see its full efffect
```
