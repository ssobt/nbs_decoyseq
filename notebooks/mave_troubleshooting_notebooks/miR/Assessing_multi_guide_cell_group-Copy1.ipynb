```python
%%R
lv_plotting_function = function(data, nm){ggplot(data, aes(LV.mu.0, LV.mu.1, color = guide)) + geom_point() + theme_classic() +
                                       coord_cartesian(xlim = c(-4,4), ylim = c(-4,4)) + ggtitle(nm) + scale_color_manual(values = c('darkblue', 'yellow', 'grey20', 'green', 'grey80'))}

umap_plotting_function = function(data, nm){ggplot(data, aes(UMAP1, UMAP2, color = guide)) + geom_point() + theme_classic() +
                                       ggtitle(nm) + scale_color_manual(values = c('darkblue', 'yellow', 'grey20', 'green', 'grey80'))}
```

```python
%%R
lv_plots = mapply(FUN = lv_plotting_function, latent_space_mtxs, meta_weight_pattern_names, SIMPLIFY = FALSE)
umap_plots = mapply(FUN = umap_plotting_function, umap_mtxs, meta_weight_pattern_names, SIMPLIFY = FALSE)
```

```python
%%R
names(lv_plots) = meta_weight_pattern_names
names(umap_plots) = meta_weight_pattern_names
```

```python
%%R
lv_plots = lv_plots[c('truly_all_even', 'all_even_high_triplet', 'all_even_low*_triplet', 'all_even_high*_triplet',
                     'truly_all_even_NoKLDloss', 'all_even_high_triplet_NoKLDloss', 
                      'all_even_low*_triplet_NoKLDloss', 'all_even_high*_triplet_NoKLDloss')]

umap_plots = umap_plots[c('truly_all_even', 'all_even_high_triplet', 'all_even_low*_triplet', 'all_even_high*_triplet',
                     'truly_all_even_NoKLDloss', 'all_even_high_triplet_NoKLDloss', 
                      'all_even_low*_triplet_NoKLDloss', 'all_even_high*_triplet_NoKLDloss')]
```
